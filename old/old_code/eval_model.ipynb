```python
import pandas as pd
import numpy as np
import glob
import lightgbm as lgb
```

```python
data_path ='./'
model_path ='../0708/model.txt'
type_list = ["上樓梯","下樓梯","大跳","小跳","前進","後退","擦白板","蹲下"]
```

```python
data = pd.read_csv('2020-07-0914.00.19.csv',usecols = np.arange(0,7)).drop_duplicates(subset = "time", keep="first")
```

```python
model = lgb.Booster(model_file=model_path)
```

```python
result = []
for idx in np.arange(0,len(data)-200,200):
    X = data[idx:idx+200]
    X = X.unstack()
    y = model.predict(X)
    print(type_list[y.argmax()])
    result.append(y)
```
